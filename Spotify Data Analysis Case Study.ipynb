```python
# Import all the necessary libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Importing the tracks dataset in Python

df_tracks = pd.read_csv('C:/Users/hasan/Documents/Python Scripts/Spotify Dataset/tracks.csv')
df_tracks.head()
```

```python
# Check for null values in dataset

pd.isnull(df_tracks).sum()
```

```python
# Information about the dataset

df_tracks.info()
```

```python
# 10 least popular songs in the dataset

sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
# Descriptive Statistics about the dataset

df_tracks.describe().transpose()
```

```python
# Top 10 most popular songs in the dataset with popularity greater than 90

most_popular = df_tracks.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False).head(10)
most_popular
```

```python
# Setting the release date column as the index
df_tracks.set_index('release_date', inplace = True)

# Changing the datatype of release date column from object to datetime
df_tracks.index = pd.to_datetime(df_tracks.index)

df_tracks.head()
```

```python
# How to get information for a particular row

df_tracks[['artists']].iloc[18]
```

```python
# Change the duration from milliseconds to seconds

df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))

# Removing the column duration_ms from the dataset
df_tracks.drop('duration_ms', inplace = True, axis = 1)

df_tracks.duration.head()
```

```python
corr_df = df_tracks.drop(['key', 'mode', 'explicit'], axis = 1).corr(method = 'pearson')
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(data = corr_df, annot = True, fmt = '.1g', vmin = -1, vmax = 1, center = 0, cmap = 'inferno', lw = 2, linecolor = 'Black')
heatmap.set_title('Correlation Heat Map between Variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90);
```

```python
# Sampling the data by only taking 0.4% of the complete data to generate regression plots
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))

print(len(sample_df))
```

```python
# Regression Plots

plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'c').set(title = 'Loudness vs Energy Correlation');
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'b').set(title = 'Popularity vs Acousticness Correlation');
```

```python
# Extracting the years from the release date column

df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
# Create a distribution plot for total songs released each year

sns.displot(data = years, discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = 'Number of songs released each year');
```

```python
# Create a barplot for Duration of songs every year

total_dur = df_tracks.duration
fig_dims = (18,7)
fig, axes = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dur, ax = axes, errwidth = False).set(title = 'Year vs Duration')
plt.xticks(rotation = 90);
```

```python
# Line plot for average duration of songs over the years

total_dur = df_tracks.duration
sns.set_style(style = "whitegrid")
fig_dims = (12,5)
fig, axes = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dur, ax = axes).set(title = 'Year vs Duration');
plt.xticks(rotation = 60);
```

```python
# Importing the Genre dataset in Python

df_genre = pd.read_csv('C:/Users/hasan/Documents/Python Scripts/Spotify Dataset/SpotifyFeatures.csv')
df_genre.head()
```

```python
# Creating a barplot for duration of songs in different genres

plt.title('Duration of the Songs in different Genres')
sns.color_palette('rocket', as_cmap = True)
sns.barplot( y = 'genre', x = 'duration_ms', data = df_genre)
plt.xlabel('Duration in ms')
plt.ylabel('Genre');
```

```python
# Top 5 genre by popularity

sns.set_style(style = "darkgrid")
plt.figure(figsize = (10,5))
famous_songs = df_genre.sort_values('popularity', ascending = False).head(10)

# Create a bar plot
sns.barplot(x = 'popularity', y = 'genre', data = famous_songs).set(title = 'Top 5 Genres by Popularity');
```
